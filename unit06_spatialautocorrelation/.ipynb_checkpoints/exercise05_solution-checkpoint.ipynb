## Data preparation

The dataset is from the parliament election in 2019 on the most detailed spatial resolution called 'afstemningsområder' (*voting areas*).

You can find the necessary data in the data subfolder. The original data source is https://valgdatabase.dst.dk/

## Task I: get the data ready!

**The data set on votes in each area contains columns with votes for each party, number of people entitled to vote etc. You have to choose yourself which part of the data you want to do your analysis on.**

```python
import seaborn as sns
import pandas as pd
import esda
from pysal.lib import weights
from splot.esda import (
    moran_scatterplot, lisa_cluster, plot_local_autocorrelation
)
import geopandas as gpd
import numpy as np
import contextily as ctx
import matplotlib.pyplot as plt
```

```python
# Read data

fp_votes = './data/electiondata.csv'
fp_geom = './data/afstemningsomraader_dk.gpkg'

votes = pd.read_csv(fp_votes, sep=';', encoding='utf-8')
geoms = gpd.read_file(fp_geom)
```

```python
votes.head(2)
```

```python
geoms = geoms[['afstemningsomraadenummer', 'afstemningsstedNavn',
       'kommunekode', 'kommuneLokalId', 'geometry']]

geoms['ValgstedId'] = geoms.kommunekode + '0' + geoms.afstemningsomraadenummer

geoms['ValgstedId'] = pd.to_numeric(geoms.ValgstedId)

geoms.head(2)
```

```python
ax = geoms.plot(facecolor='purple', edgecolor='black', alpha=0.5, figsize=(15,15), linewidth=0.3)

ctx.add_basemap(ax, crs=geoms.crs)
```

```python
# Check crs
geoms.crs
```

```python
v_areas = geoms.join(votes.set_index('ValgstedId'), on='ValgstedId', how='inner')

v_areas.drop(['Gruppe','kommuneLokalId','afstemningsomraadenummer', 'Gruppe', 'KredsNr',
            'StorKredsNr', 'LandsdelsNr','FV2019 - 1. Uden for partierne',
       'FV2019 - 2. Uden for partierne', 'FV2019 - 3. Uden for partierne',
       'FV2019 - 4. Uden for partierne', 'FV2019 - 5. Uden for partierne'], axis=1, inplace=True)


v_areas.head(2)
```

```python
v_areas['pct_enhed'] = v_areas['FV2019 - Ø. Enhedslisten - De Rød-Grønne'] / v_areas['FV2019 - Afgivne stemmer'] * 100
```

```python
# Create weights and check for islands

wk = weights.distance.Kernel.from_dataframe(v_areas, fixed=False, k=15)

assert len (wk.islands) == 0
```

```python
# Standardize spatial weights matrix
wk.transform = 'R' # Using row standardization

print(round(sum(wk[0].values()),9)) # Now the sum of each feature's weights should sum to one
 
```

```python
# Create standardized version of vote counts 
v_areas['pct_enhed_std'] = (
    v_areas['pct_enhed'] - v_areas['pct_enhed'].mean()
) / v_areas['pct_enhed'].std()
```

```python
# Create spatial lag of case numbers
v_areas['w_pct_enhed_std'] = weights.lag_spatial(wk, v_areas['pct_enhed_std'])

v_areas[['pct_enhed_std','w_pct_enhed_std']].head()
```

```python
v_areas.plot(column='pct_enhed_std', scheme='equal_interval', k=10, cmap='magma', legend=True, figsize=(15,15))
```

## Task II: global spatial autocorrelation

Let's move on to the analytics:

- Visualise your data score with a Moran Plot
- Calculate Moran's I
- *What conclusions can you reach from the Moran Plot and Moran's I? What's the main spatial pattern?*

```python
# Visualise with a Moran Plot

f, ax = plt.subplots(1, figsize=(9, 9))

sns.regplot(x='pct_enhed_std', y='w_pct_enhed_std', data=v_areas, ci=None)

plt.axvline(0, c='k', alpha=0.5)
plt.axhline(0, c='k', alpha=0.5)

# OBS! The placement of the text is customized to this specific plot and the range of the axes
plt.text(3, 1.5, "HH", fontsize=25)
plt.text(3, -0.7, "HL", fontsize=25)
plt.text(-1, 2, "LH", fontsize=25)
plt.text(-0.9, -1.2, "LL", fontsize=25)


plt.show()

```

```python
# Calculate Moran's I

mi = esda.Moran(v_areas['pct_enhed'], wk)

print("Moran's I:", mi.I)

print("P-value for the Moran's I statistic:", mi.p_sim)

moran_scatterplot(mi)
```

## Interpretation

Our plots displays a positive relationship between the two variables (i.e. between the standardized value for percent votes for the party and the weights, which can be interpreted as the average of the neighbour's voting behaviours).
The relationship is not perfectly linear, but looking at the Moran's I and the p-value for Moran's I we can nevertheless reject that there is a stronger pattern with similar values being close together, than if the values had been randomly distributed.

Areas with many votes for this party thus tends to be surrounded by other areas which also have many votes for this party, and vice versa.
We thus has a case of (global) spatial autocorrelation.


## Task III: local spatial autocorrelation

Now that you have a good sense of the overall pattern in the dataset, let's move to the local scale:

- Calculate LISA (Local Indicators of Spatial Association) statistics for the voting areas.
- Make a map of significant clusters at the 5%
- Can you identify hotspots or coldspots? If so, what do they mean? What about spatial outliers?


The last part is a bit more sophisticated:

- Create cluster maps for significance levels 1% and 10%; compare them with the one we obtained. *What are the main changes? Why?*

```python
# LISA statistics

lisa = esda.Moran_Local(v_areas['pct_enhed'], wk)
```

```python
# Map of significant clusters at 5 %

# Break observations into significant or not
v_areas['significant'] = lisa.p_sim < 0.05
# Store the quadrant they belong to
v_areas['quadrant'] = lisa.q

v_areas['significant'].head()
```

```python
lisa.p_sim[:5]
```

```python
v_areas.quadrant.head()
```

```python
fig, ax = plt.subplots(1, figsize=(10,10))

lisa_cluster(lisa, v_areas, ax=ax)
```

```python
plot_local_autocorrelation(lisa, v_areas, 'pct_enhed');
```

## Interpretation
The LISA map is in a way a simplified choropleth map of the values - it shows a simplifed and generalized picture of where there are respectively areas with clusters of high or low values.

For the percentage used here we see clusters of HH (hotspots) in the larger cities and around south of Funen. Meanwhile some of the areas south of Cph belong to a LL cluster - low values surronded by other low values (coldspots).
Finally, there are a few HL areas in Jutland - i.e. voting areas with a high value surrounded by low values (outliers).

```python
# Cluster maps for significance levels 1% and 10%

# Break observations into significant or not
v_areas['significant_01'] = lisa.p_sim < 0.01
v_areas['significant_10'] = lisa.p_sim < 0.1

v_areas[['significant_01','significant_10']].head(30)
```

```python
fig, ax = plt.subplots(1, figsize=(10,10))

lisa_cluster(lisa, v_areas, 0.01, ax=ax)

ax.set_title('LISA cluster at significance level 0.01')

plt.show()
```

```python
fig, ax = plt.subplots(1, figsize=(10,10))

lisa_cluster(lisa, v_areas, 0.1, ax=ax)

ax.set_title('LISA cluster at significance level 0.1')

plt.show()
```

## Interpretation

The higher the significance threshold (meaning our tolerance for the likelihood of getting the same spatial patterns randomly), the fewer clusters are found, and vice versa.
